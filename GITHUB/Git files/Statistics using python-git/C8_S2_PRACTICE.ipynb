```python
def outliers(col):
    ser=pd.Series(df[col])
    q1=ser.quantile(0.25)
    q3=ser.quantile(0.75)
    iqr=q3-q1
    uf=q3+1.5*iqr
    lf=q1-1.5*iqr
    for z in ser:
        if z>uf or z<lf:
            df[col]=df[col].replace([z],df[col].mean())
```

```python
#TASK3
#10 15 30 200 1000 5000 10000 50000 100000
num = [10,15,30,200,1000,5000,10000,50000,100000]
data_s = []
fig,ax = plt.subplots(3,3,figsize=(14,10))
for i in num:
    sample_df=fare_amount.sample(n=i,replace=True,random_state=1)
    data_s.append(sample_df['fare_amount'])
k=0
for i in range(0,3):
    for j in range(0,3):
        ax[i,j].hist(data_s[k],bins=10,density=True)
        ax[i,j].set_title(label='Sample_size='+str(len(data_s[k])))
        k=k+1
plt.show()        
```

#interpretation
For Sample Sizes 200 and Above (200, 1000, 5000, 10000, 50000, 100000):

As the sample size increases, the distribution becomes smoother and more closely resembles a normal distribution due to the Central Limit Theorem.
By the time the sample size reaches 1000 or more, the distribution is likely nearly normal, even if the population itself isnâ€™t perfectly normal.
Larger sample sizes reduce the impact of random variability and provide a more accurate representation of the population.

```python
#TASK5
sample_df1 = pd.DataFrame()
for i in range(0,20):
    sample_df1['sample_' +str(i)] = df['fare_amount'].sample(n=200, replace = True).reset_index(drop = True)
sample_df1
```

```python
sample_mean = []
sample_name = sample_df1.columns
for i in sample_name:
    sample_mean.append(sample_df1[i].mean())

sample_mean_s = pd.DataFrame()
sample_mean_s.insert(0,'sample_name',sample_name)
sample_mean_s.insert(1,'sample_mean',sample_mean)
print(sample_mean_s)
```

```python
import statistics as st
import scipy.stats as sts
from scipy.stats import kurtosis
```

```python
#population parameters
mean_p = df['fare_amount'].mean()
print('Population mean is:',mean_p)
median_p = df['fare_amount'].median()
print('Population median is:',median_p)
std_p = st.stdev(df['fare_amount'])
print('Population std is:',std_p)
skew_p = sts.skew((df['fare_amount']), bias = False)
print('Population skewness is:',skew_p)
kurtosis_p = kurtosis((df['fare_amount']), bias = False)
print('Population kurtosis is:',kurtosis_p )
```

```python
sample200 = df[['fare_amount']].sample(n=200 , replace = True, random_state=1,ignore_index=True)
sample200
```

```python
#sample200 parameters
mean_s = st.mean(sample200['fare_amount'])
print("sample200 mean is:", mean_s)
median_s = st.median(sample200['fare_amount'])
print("sample200 median is:", median_s)
std_s = st.stdev(sample200['fare_amount'])
print("sample200 std is:", std_s)
skew_s = sts.skew(sample200['fare_amount'], bias = False)
print('sample200 skewness is:',skew_s)
kurtosis_s = kurtosis(sample200['fare_amount'], bias = False)
print('sample200 kurtosis is:',kurtosis_s)
```

Mean & Median:
The average (mean) and middle value (median) are nearly the same for both the population and the sample of 200.
Kurtosis & Skewness:
The sample has some bias and is slightly stretched to the right (positively skewed).
The population has lighter tails (kurtosis ~ 0.6) and is also slightly stretched to the right.
Standard Deviation:
The variation in the sample mean (from 20 numbers) is about the same as the sample's standard deviation divided by the square root of the sample size.
Shape of Distribution:
The population and the sample of 200 donâ€™t look like a normal (bell-shaped) curve and have lighter tails. The sampleâ€™s shape is somewhat similar to the population, but not exactly the same.
